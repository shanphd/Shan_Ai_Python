```python
print(bool('123'))
print(bool(12))
print(bool(""))
print(bool())
print(bool(None))
print(bool("None"))
```

```python
# any() usecase out of 3 varibale one varibale shoulr be filled

#True

name = ''
age = 37
email = " "

print(any([name,age,email]))

#False
name = ""
age = ""
email = ""

print(any([name,age,email]))
```

```python
# all() usecase all 3 varibale should be filled

#True

name = 'shan'
age = 37
email = "jk@gmail.com"

print(all([name,age,email]))

#False

name = 'shan'
age = ""
email = "jk@gmail.com"
print(all([name,age,email]))

```
