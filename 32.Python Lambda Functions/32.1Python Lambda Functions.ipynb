```python
#Lambda 

add = lambda x, y: x+y
print(add(2,3))
```

```python
check = lambda i: i in "python"
print(check('n'))
```

```python
# Custom Lambda function

prices = ['$123.13','$230.78','$89.09']
print(type(prices))
#Need to remove $
p = '$123.13'
print(float(p.replace('$', "")))
print(type(p))
```

```python
prices = ['$123.13','$230.78','$89.09']
s =(list(map(lambda p: float(p.replace('$', "")),prices)))
print(s)
avg = sum(s)/len(s)
print(avg)
```

```python
student = [['Max',100],
           ['shan',80],
           ['wen',70]]

print(student)
print(list(filter(lambda row : row[1]>70,student)))
print(list(filter(lambda row : row[0].startswith('M'),student)))
```
