### List clear,remove,pop



```python
from IPython.display import Image,display
display(Image("27.4.png",width=400))
display(Image("27.5.png",width=400))
```

```python
matrix_name = [
    ["a","r","e","w","n","a","l"],  #0 Row
    ["c","h","a","r","m","p"],      #1 Row
    ["e","n","l","a","e","n","d"]   #2 Row
]
#print(matrix_name)
```

```python
matrix_name[0].clear()
print(matrix_name)
```

```python
matrix_name = [
    ["a","r","e","w","n","a","l"],  #0 Row
    ["c","h","a","r","m","p"],      #1 Row
    ["e","n","l","a","e","n","d"]   #2 Row
] 

matrix_name[1].remove('r')
print(matrix_name)
```

```python
matrix_name = [
    ["a","r","e","w","n","a","l"],  #0 Row
    ["c","h","a","r","m","p"],      #1 Row
    ["e","n","l","a","e","n","d"]   #2 Row
] 

n = matrix_name[2].pop(4)
print(matrix_name)
print(f"Removed item:", n)
```
