### Sort


```python
matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l']
]
print(matrix_name)
```

```python
matrix_name.sort()
print(matrix_name)
```

```python
# to sort only row 1
matrix_name[0].sort()
print(matrix_name)
```

```python
matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l']
]
print(matrix_name)
```

```python
matrix_name[1].sort(reverse=True)
print(matrix_name)
```

```python
matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l']
]
print(matrix_name) 
```

```python
matrix_name[1].sort(reverse=True)
print(matrix_name)
```

```python
matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l']
]
print(f"OLD MATRIX NAME : {matrix_name}")
```

```python
new_matrix_name=sorted(matrix_name)
print(f"NEW MATRIX NAME: {new_matrix_name}")
```

```python
new_matrix_name=sorted(matrix_name[1], reverse=True)
print(f"NEW MATRIX NAME: {new_matrix_name}")
```

```python
### REVERSE

matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l'],
    ['b','o','b']
] 

matrix_name.reverse()
print(matrix_name)
```

```python
matrix_name = [
    ['s','h','a','n'],
    ['a','r','s','e','n','a','l'],
    ['b','o','b']
] 

matrix_name.reverse()
print(matrix_name)
```
