```python
# List Comprehension Example 

domain = ["shan.com","WWW.openai.com","localhost",'GooGLE.com']
print(f"#"*20)
print(domain)
print(f"#"*20)
```

```python
cleaned = [
    d.lower().replace('www.', '')
    for d in domain
    if '.' in d
]
print(cleaned)
```

```python
price = [80,30,100]
mul = [
    p*2
    for p in price
    if p > 50 
    ]
print(mul)
```
