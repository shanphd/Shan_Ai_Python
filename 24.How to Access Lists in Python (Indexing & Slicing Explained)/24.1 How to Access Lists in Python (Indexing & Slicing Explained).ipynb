```python
# Nested 
matrix = [
    ['a','b','c'],
    [1,9,90],
    ['w','e','f']
    ]

for row in matrix:
    for item in row:
        print(item, end=" ")
    print()

print(matrix[1][-1])
print(matrix[0][1])


       
```
