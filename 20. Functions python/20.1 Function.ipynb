### 20.1 Function in python

```python
# without parameter 

def greeting_message():
    print(f"Welcome to python world")
    
greeting_message()
```

```python
# With parameter

def greeting_message(name):
    print(f" Hello {name} welcome to python world")
```

```python
greeting_message('shan')
```

```python
def epl_champs(clubname,yearwon,captain_name,coach_name):
    print(f"Epl_champ_clubname: {clubname}")
    print(f"Epl_champ_yearwon: {yearwon}")
    print(f"Epl_champ_captainname: {captain_name}")
    print(f"Epl_champ_coach_name: {coach_name}")
    
    
```

```python
epl_champs("Arsenal",2026,"odegar","Arteta")
```

```python
# Default argument

def shan_pc (year,pc='shan'):
    print(f"The pc belong to {pc} bought in year{year}")
```

```python
shan_pc(2026,"lol")
```

```python
shan_pc(2026)
```

```python
# Keyword arg 

def audit(id,name,amt,year):
    print(f"Transaction id number : {id}")
    print(f"name of user : {name}")
    print (f"Amount of user : {amt}")
    print(f"yearof amount trancasation : {year}")
```

```python
audit(id=90,name='shan',amt=2000000,year=2026)
```

```python
## Variable arg 

def sales(*amount):
    print(f"Total amount : {amount}")
    print(f"sales of amount: {sum(amount)}")
```

```python
sales(100,200,200)
```

```python
## Keyword arg it is dictwith key value

def emp_pro(**details):
    for key, value in details.items():
        print(key, ":", value)
```

```python
emp_pro(name = "shan", id = '241')
```
