### Python relationship

```python
dtcc_profile = {
    'name':'shan',
    'age':37,
    'team':'snowflake',
    'seat_no':'che03',
    'row':4,
    }

print(f"{'#'*20}")

# Access 
print(dtcc_profile.get('age'))
print(f"Dictonary access : {dtcc_profile.get('name')}")
print(f"Dictonary access : {dtcc_profile.get('sysid','unknown')}")

# check 

print(f"Dictonary check : {'name' in dtcc_profile }")
print(f"Dictonary check : {'sysid' in dtcc_profile }")

#view objects 

print(f"Dictonary obkects : {dtcc_profile.items()}")
```

```python
# for loops 

for key,values in dtcc_profile.items():
    print(f"{key}:{values}")
```

```python
# Dic nice mwthod 

dtcc_id = dict.fromkeys(['id','name','age'],None)
print(f"Python dictonary tip method: {dtcc_id}")
dtcc_id['age']=40
print(f"Python dictonary tip method: {dtcc_id}")
dtcc_id.update(name='shan',id=100)
print(f"Python dictonary tip method: {dtcc_id}")
```

```python
# keep only string value and change value uppercase 

dtcc_profile = {
    'name':'shan',
    'age':37,
    'team':'snowflake',
    'seat_no':'che03',
    'row':4,
    }

#
dtcc_str = {
    k: v.upper()
    for k, v in dtcc_profile.items()
        if isinstance (v,str)
        }
print(dtcc_str)
    
```
