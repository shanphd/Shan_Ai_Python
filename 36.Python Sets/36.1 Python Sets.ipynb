```python
## Set Example 

sys_id = {101,102,103,101,109}
print(f"set is unordered: {sys_id}")
print(f"set is non duplicate: {sys_id}")
sys_id1 = {101,102,103,101,109}
sys_id1.remove(102)
print(f"set is non mutable: {sys_id1}")
```

```python
# Set ADD 
sys_id = {101,102,103,104,109}
sys_id.add(107)
sys_id.add(110)
print(f"Set add function: {sys_id}")
```

```python
# Set update 
sys_id = {101,102,103,104,109}
sys_id |= {100,99,98}
sys_id.update("shan")
print(f"Set update function: {sys_id}")
```

```python
# set remove 
sys_id = {101,102,103,104,109}
sys_id.remove(109)
sys_id.discard(100) # no error
print(f"Set remove function: {sys_id}")
```

```python
# Set union 
sys_id1 = {'sfk','dwc','itp','mdu','dsc'}
sys_id2 = {'sfk','dda','rrp','mde','dsu'}

#union 
sys_id_un = sys_id1 | sys_id2
#diff
sys_id_diff = sys_id1 - sys_id2
#symmentric
sys_id_symm = sys_id1 ^ sys_id2
#intersetion 
sys_id_int = sys_id1 & sys_id2
print(f"set union output: {sys_id_un}")
print(f"set diff output: {sys_id_diff}")
print(f"set symm output: {sys_id_symm}")
print(f"set int output: {sys_id_int}")
```
