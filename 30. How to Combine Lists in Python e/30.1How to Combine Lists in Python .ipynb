```python
# Combination Method 

name = ['s','h','a','n']
number = [1,2,3]

# Normal Addition 

print(f"Add Combination : {name} + {number}")

comb = name + number
print(comb)
```

```python
name = ['s','h','a','n']
num = [1,2,3]
com1 = [name,num]
print(com1)
```

```python
# Extend 

name = ['s','h','a','n']
num = [1,2,3]

ext = num.extend(name)
print(num)
```

```python
### ZIP Combination

ids = [101,102,103]
name = ['shan','john','Tom']
emp_id = zip(ids,name)
print(emp_id)
print(f"The emplyoee id is :", list(zip(ids,name)))
```
