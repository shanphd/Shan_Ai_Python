```python
### Iterator through Loops 

name = ['s','h','a','n']
for n in name:
    print(n)
```

```python
### iterator adding a new list and transform 

name=['s','h','a','n']
new_name=[]
for n in name:
    new_name.append((n).upper())
    print(new_name)
```

```python
### Enumerate 

name = ['d','t','c','c']
print(enumerate(name))
```

```python
name = ['d','t','c','c']
print(list(enumerate(name)))
```

```python
name = ['d','t','c','c']
for index,value in enumerate(name):
    print(index,value)
```

```python
## Reverese 

name = ['a','b','c']
for n in reversed(name):
    print(n)
```

```python
## Combine 

name = ['a','b','c','d']
num = ['1','2','3']
for l,n in zip(name,num):
    print(l,n)
```

```python
# convert str into int using map 

num = ['2','3','4']
print(list(map(int, num)))
print(type(num))
```

```python
name = ['jo ', ' ko l', 'opp   '] 
print(list(map(str.strip,name)))
```

```python
name = ['jo ', ' ko l', 'opp   '] 
for n in (map(str.strip,name)):
    print(n)
```
