```python
# List function 

sfk_sysid = ['dwc','sfk','mde']
print(f"{"#"*10}ordered{'#'*10}")
print(sfk_sysid)
print(f"{"#"*10}duplicate{'#'*10}")
sfk_sysid1 = ['dwc','sfk','mde','sfk']
print(sfk_sysid1)
print(f"{"#"*10}Indexed{'#'*10}")
sfk_sysid2 = ['dwc','sfk','mde','sfk']
print(f"Index",list(enumerate(sfk_sysid2)))
print(f"{"#"*10}Mutable{'#'*10}")
sfk_sysid3 = ['dwc','sfk','mde','sfk']
sfk_sysid3[3]='mrd'
print(f"Mutable",list(enumerate(sfk_sysid3)))
```
