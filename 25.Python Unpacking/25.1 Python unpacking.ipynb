### Asterik *

```python
person = ['shan',36,'Data scientist','Chennai','India']

name,*details,country = person 
print(name)
print(details)
print(country)
```

```python
person = ['shan',36,'Data scientist','Chennai','India'] 

*details,country = person

print(details)
print(country)
```

### _ underscore

```python
epl= ['Arsenal','martin','captain','England','2026']

print(epl)

name,*_,year = epl # *_ is like deleting to avaoid space in python 
print(id(name))
print(year)
print(_)
print(id(_)) 
```
