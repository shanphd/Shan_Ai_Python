```python
# Double Quote and Single Quote

name = 'shan'
role = 'Ai Engineer'

print(f"{name}{role}")
```

```python
name = 'shan'
role = 'Ai Engineer'

print(f"{name} \'{role}'")
```

```python
name = 'shan'
role = 'Ai Engineer'

print(f"{name} \\ {role}")
```

```python
name = 'shan'
role = 'Ai Engineer'

print(f"{name}\n{role}")
```

```python
name = 'shan'
role = 'Ai Engineer'

print(f"{name}\t{role}")
```
