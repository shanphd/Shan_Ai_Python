```python
score = [85,90,70]
print(f"Max num :", max(score))
print(f"Min num :", min(score))
print(f"Sum num :", sum(score))
print(f"All num :", all(score))
print(f"All num :", all([1,0,2]))
print(f"Any num :", any([1,0,2]))
print(f"Any num :", any([0,0,0]))
print(f"Len num :", len([0,0,0]))
```

```python
letter = ['a','c','a','d']
print(f"count fn:", letter.count('a'))
print(f"Index fn:", letter.index('d'))
print(f"is fn:", 'a' in letter)
print(f"is fn:", 'a' not in letter)
```

```python
list1 = ['1','3','5']
list2 = ['5','3','5']

print(f"Equal fn:", list1 == list2)
print(f"is fn:", list1 is list2)
print(f"less fn:", list1 < list2)
```
