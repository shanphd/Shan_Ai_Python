### 19.1 For vs While Loops

```python
# for loops

for i in range (5):
    print(i)
```

```python
# for loops with DYnamic change Nested loops

years = ['2026', '2027']
months = ['Jan','Feb']
dates = range (1,9)

for year in years:
    for month in months:
        for date in dates:
            print(f"snowflake_wh_report.{year}_{month}_{date}.csv")
```

```python
### While loops 
i = 0
while i <= 4:
    #print(i)
    i += 1
    print(i)
    
```

```python
### While True statement 

answer = "" 
attempts = 0

while True:
    if attempts < 3:
        answer = input ("Do you agree : yes/no ?")
        attempts += 1
        
        if answer == 'yes':
            print (f"user answer : {answer}")
            break
    else:
        print("You mentioned different answer 3 attempt over")
        break
```
