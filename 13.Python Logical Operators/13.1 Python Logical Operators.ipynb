### AND,OR usecase 
Comparing two value with chained operator 


```python
# OR usecase to check cpu_usgae and Memory_usage 

def email_alert_pchigh(cpu,mem):
    if cpu  >= 90 or mem >= 90:
        return "Email alert snowflake pc is high"
    elif cpu >= 80 or  mem >= 80:
        return "Email alert snowflake pc is about to high"
    else:
        return "Email alert snowflake pc is normal"
```

```python
print(email_alert_pchigh(70,70))
print(email_alert_pchigh(90,70))
print(email_alert_pchigh(80,70))
```
