```python
# Shallow Copy 

name = ['a','r','s','e','n']
new_name = name.copy() # -> shallow copy 
new_name.append('x')
new_name.insert(0,'a')
print(f"old name : {name}")
print(f"New name : {new_name}")
```

```python
epl= [
    ['m','a','n'],
    ['a','r','s']
       ]
       
epl_new = epl.copy()

epl[0].append('a')
epl_new[1].append('l')
print(f"org: {epl}" )
print(f"copy: {epl}" )
```

```python
### Method Import Copy 

import copy 

dtcc = [
    ['s','f','k'],
    ['d','w','c']]

dtcc_new = copy.deepcopy(dtcc)

dtcc[1].append('y')
dtcc_new[0].append('z')

print(f"org: {dtcc}")
print(f"copy: {dtcc_new}")
```

```python
### How to check Copy Function 

import copy 
sysid = [
    ['s','f','k'],
    ['m','d','u']
] 

sysid_new = sysid.copy()
#copy()

print("shared_object", sysid == sysid_new, "\n")

#shallowCopy 

sysid1_new = copy.deepcopy(sysid)

print("shared_object", sysid == sysid1_new, "\n")
print("shared1_object", sysid[0] == sysid1_new[1], "\n")
```
